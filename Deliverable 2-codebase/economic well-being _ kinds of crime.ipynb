# Data (Graphs & Infographics)
The following infographics shows several datas related to economic well-being, familty/ community and safety/risky behaviors

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("Labor Force, Employment, Unemployment:rate ward6.xlsx")

print(var)
var.plot.line(title='Labor Force, Employment, Unemployment, in ward6 in DC',x='Year', y=['Labor Force','Employment', 'Unemployment'], grid=True)
var.plot.bar(title='Labor Force, Employment, Unemployment in ward6 in DC',x='Year', y=['Labor Force','Employment', 'Unemployment'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("Labor Force, Employment, Unemployment:rate ward6.xlsx")

print(var)
var.plot.line(title='Unemployment rate in ward6 in DC',x='Year', y='Unemployment rate', grid=True)
var.plot.bar(title='Unemployment rate in ward6 in DC',x='Year', y='Unemployment rate')
```

The unemployment rate measures the share of workers in the labor force who do not currently have a job but are actively looking for work.

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("Median_income_of_families_with_children_in_ward6.xlsx")

print(var)
var.plot.line(title='Median_income_of_families_with_children_in_ward6',x='TimeFrame', y='MedianAnnualIncome', grid=True)
```

The graph above shows:

(Currency)Median annual income, in inflation-adjusted dollars, of families with children age 0-17.

For households and families, the median income is based on the distribution of the total number of households and families inclusing those with no income.

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("Children in single-parent families in ward6 in DC.xlsx")

print(var)
var.plot.line(title='Children in single-parent families in ward6 in DC',x='Year', y='Percentage', grid=True)
var.plot.bar(title='Children in single-parent families in ward6 in DC',x='Year', y='Percentage')
```

The graph above shows:

Children under age 18 who live with their own single parent either in a family or subfamily.

In this definition, single-parent families may include cohabiting couples and do not include children living with married stepparents. Children who live in group quarters (for example, institutions, dormitories, or group homes) are not included in this calculation.

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("Substantiated investigations of child neglect and abuse in ward6.xlsx")

print(var)
var.plot.line(title='Substantiated investigations of child neglect and abuse in ward6',x='TimeFrame', y='Data', grid=True)
var.plot.bar(title='Substantiated investigations of child neglect and abuse in ward6',x='TimeFrame', y='Data')
```

The graph above shows:

Child abuse/neglect: Physical or mental injury, sexual abuse, negligent maltreatment, or maltreatment of any child under age 18 by a person responsible for the child's welfare under circumstances that threaten or harm the child's health or safety.

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("teens ages 16 to 19 not attending school and not working in DC.xlsx")

print(var)
var.plot.line(title='teens ages 16 to 19 not attending school and not working in DC',x='Year', y='Number', grid=True)
var.plot.bar(title='teens ages 16 to 19 not attending school and not working in DC',x='Year', y='Number')
```

The graph above shows:

Teenagers between age 16 and 19 who are not enrolled in school (full- or part-time) and not employed (full- or part-time).

This measure is sometimes referred to as “Idle Teens” or “Disconnected Youth.”
