# Population and Age Trends in DC Ward 6


```python
import plotly.express as px

info = "Considering the total population from past 5 years the below graph is ploted, It shows that Tract 102.02 is the most populated Tract in the past 5 years"
print(info)
print("\n")

# Data for the plot
years = [
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"]
populations = [2522,1960,4027,2385,1519,2492]

# Create a bar plot using Plotly
fig = px.bar(x=years, y=populations, color=years, color_discrete_sequence = px.colors.sequential.Plasma_r)
#color_discrete_sequence=px.colors.qualitative.Light24

# Set the plot title and axis labels
fig.update_layout(
    title={
        'text': "Total Population Tract wise - ACS 5 Years",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title="Year",
    yaxis_title="Total Population")

# Show the plot
fig.show()
```

## The below Age trends graphs are plotted for different Tracts under ward 6 - southwest DC. Each bar chart shows population trends of different age groups.")

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def renderIncomeData(excelName, sheetName, tractName):
    df = pd.read_excel(excelName, sheet_name=sheetName)
    df = df.dropna()
    
    fig, ax = plt.subplots(figsize=(20,20))

    ind = np.arange(len(df))
    width = 0.4
    ax.barh(ind, df["Total"], label = "Total", color = "plum")
    ax.barh(ind, df["Male"], label = "Male",  color = "red")
    ax.barh(ind, df["Female"], label = "Female", color = "purple")
#   ax.barh(ind, df["Nonfamily households"],label = "Nonfamily households")
    ax.set(yticks=ind, yticklabels=df["Label"])

    # df = pd.DataFrame({'Total': df["Total"], 'Below Poverty Level': df["Below poverty level"]}, index=df["Label (Grouping)"])
    # ax = df.plot.barh()

    plt.title('Population Trends for different ages ' + tractName, fontsize=14, weight='bold', alpha=.8)
    plt.xlabel('Grouping', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    ax.legend()
    plt.show()

excelName = "./dataset/ageTrends_Tractwise.xlsx"
sheetNames = [
    "DC",
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"]

for sheetName in sheetNames:
    renderIncomeData(excelName, sheetName, sheetName)
```
